## Week1:What is AI?
#### Introduction:

**AI:**
* ANI(artificial narrow intelligence):  
  人工狹義智能，這些AI只能做一件事，例，自動駕駛汽車、工廠中的AI應用程序。
* Generative AI(generative artificial intelligence):  
  生成式人工智慧，例如GPT可以幫助你編輯文案、提出建議、生成摘要...等。
* AGI(artificial general intelligence):  
  人工通用智能，為建構人工智能的目標，他可以完成人類所能完成的任何智力任務，甚至有可能比人類變的更加聰明。

**大型語言模型，large language models (LLMs):**  
從大量文本中學習，透過反覆預測該輸出的下一個單詞來生成新的文本。  
  | Input(A) | Output(B) |
  | :-- | :-- |
  | My favorite drink | is |
  | My favorite drink is | lychee |
  | My favorite drink is lychee | bubble |
  | My favorite drink is lychee bubble | tea |
  → My favorite drink is lychee bubble tea

* 傳統的人工智能系統，在最初他的表現會隨著提供更多數據而有更好的表現，但在某個程度之後，表現增加幅度會逐漸平緩，(儘管輸入更多數據，但得到的輸出沒有變的更加準確)  
* 而隨著神經網路大小的提高，展現出的性能也會越來越好  
* 由此可知，擁有大量數據對於提高神經網路的表現確實有幫助，這也是大數據(big data)存在的原因。  
* 再者，若想訓練出一個更大的神經網路，處理器(ex.graphics processor units、GPU)超級電腦、摩爾定律(Moore's law)等等的發展佔據重要地位。

**獲取Data方式**  
* 手動標記(manual labeling):
  可同時取得具有A和B的資料集。
* 觀察行為(observing behaviors):
  ex.觀察消費者在自己的購物網站中的購買情形，建立用戶ID、登入時間、商品品項、價格、是否購買的資料集；工廠中機器的故障情形，機器編號、溫度、壓力、是否故障，用以預防未來機器故障情形的發生。
* 從網站、合作夥伴中獲取資料  

**Use and mis-use of data**  
1. 不建議收集完所有資料後再開始進行AI訓練，應該在收集資料的過程中就馬上將資料提供給AI團隊，因為這樣AI團隊才可以立即反應給你應該注重於收集哪些類型的資料，且應該在蒐集資料的過程中做出哪些改變，幫助提高AI的訓練成效。  
2. 數據並非總是具有價值的。  

**Data is messy**  
* Garbage in, garbage out:  
  數據很混亂，若含有太多錯誤的數據，那AI可能學不到東西。  
* Data problems:
  * Incorrect labels
  * Missing values  
  AI團隊必須知道如何處理錯誤的標籤、缺失的值，
* Multiple types of data
  * ex. 非結構化數據：image, audio, text  
    現今的人工智慧主要用於產生非結構化數據，而監督學習可以非常有效地處理非結構化資料和結構化資料。

### More examples of what machine learning can and cannot do  
* Self-driving car:  
  * 可以達成:分析車輛面前的圖片，可能是圖片再加上雷達或其他傳感器，再獲得希望的輸出。
  * 無法達成:分析人類手勢的意圖，因為人的手勢可以有很多不同的動作，很難從成千上萬個仁中獲取所有不同的手勢樣本，且對於車輛而言，分析手勢需要高度精確，以避免意外的發生。
    
* X-ray diagnosis:  
  * 可以達成:從大約10000張中被標記的影像中學習，進而診斷出肺炎。
  * 無法達成:從10張影像以及閱讀醫學書籍關於肺炎的片段，診斷出肺炎，而人類可以。
      

**機器學習的優點跟缺點**  
* 優點:ML擅長學習簡單、具有大量數據的東西。  
* 缺點:ML不擅長從少量數據中學習複雜概念；用系統從未見過的數據類型去執行任務，例如照片相較訓練時給的資料處於一個奇怪的角度，人工智慧不擅長舉一反三的能力會導致錯誤的輸出。

* 對業務有用的事情，由領域專家分析。

**Brainstorming framework**  
* 考慮任務自動化比自動化工作有效，從自己的業務中選出可以交給機器處理的任務。
* 商業價值的主要驅動力是甚麼?

### Working with an AI team  

**設置你的接受標準(Specify your acceptance criteria)**  
人工智能系統的特點是，性能表現通過統計數據的方式來展現。  
ex. 若目標是95%的精確度，AI團隊需要有一組資料集去測量模型表現，如此才可以知道模型是否達到了95%的精確度，而這組數據集為稱為，**數據集(test set)**，數據集可以不用太大，數據集的大小可以詢問專業人士，以達到更精準的數字。  

* 訓練集(Train set):  
  一堆配有標籤的圖片，給予了輸入A，同時還包含了輸出B。
* 數據集(Test set):  
  不配有標籤，將資料輸入訓練過的AI後檢查他會給出什麼輸出。  
通常訓練集會比數據集大的多，而在不同的項目中，有時候會需要兩個Test set。

精確度低的原因:
1. 機器學習的限制
2. 資料集太少
3. 數據很混亂。ex.數據可能被錯誤的標記
4. 模糊的標籤。ex.有些數據給人類做檢測時可能會得到不同的結論。

### CPU vs. GPU  
* CPU:Computer processor (Central Processing Unit)  
  電腦中的電腦處理器，代表中央處理單元，會在計算機中進行大量計算。
* GPU:Graphics Processing Unit  
  圖形處理單元，以往，GPU是為了處理圖片而設計的，之後發現GPU在建構大型神經網路或大型深度學習算法都有很好的表現，具有強大的計算能力去訓練越來越大的神經網路。  

### Cloud vs. On-premises  
* Cloud:  
  租用計算服務器，以便使用他人的服務進行計算。
* On-premises:  
  本地佈署指購買自己的計算服務，並在自己的公司本地運行該服務。
* Edge deployments:  
  邊緣佈署，計算在本地進行，將處理器設置再收集數據的地方，優點是可以減少系統回應的時間、減少利用網路發送的數據量。ex.自動駕駛汽車沒有足夠多的時間將數據傳到雲端再等待傳回的訊息，去決定當下是否應該停車。

### Case study  
* Smart speaker:  
  處理指令需要的步驟:
  1. 探測喚醒詞(Trigger word/wakeword detection):  
     一旦聽到喚醒詞他的輸出就是0或1，若成功喚醒接著執行第二步。
  2. 語音識別(Speech recognition):  
     喚醒後，依據喚醒詞後的句子使用另一個映射去得到喚醒詞後句子的文本。
  3. 意圖識別(Intent recognition):  
     使軟體明白你希望軟體做什麼事情，此時需要使用另一個映射，輸入轉換的文本去得到你的意圖。ex.說笑話、放音樂...
  4. 執行(Execute):  
     得到意圖後，假設你希望AI語音系統說一個笑話，那他就會執行編寫好的代碼中隨機的一個笑話。  
  這幾個步驟被稱為人工智能管道(AI pipeline)。

* Self-driving car:  
  決定如何開車需要的步驟:  
  
  1. Car detection 
  2. Pedestrian detection 
  3. Motion planning
     
    
如上方的例子，可知道需要一整個團隊才能建構出一個複雜的AI產品。

### AI轉型步驟  
1. 實施試點計畫以獲得動力  
  * 在考慮初期的項目時，重點應該放在能否成功，而非是否最有價值；先透過成功的案例吸引關注，使的他人希望與之合作。
  * 能在6-12個月之間就可展示成效的項目。
  * 這個試點可以外包，也可以交由公司內部進行，但最終還是需要擁有自己的AI團隊。
2. 建立公司內部的AI團隊
  * 先搭建一個集中的AI團隊，在從這個團隊中派人去幫助其他部門(稱為，matrix organization)
  * AI團隊需要搭建一個全公司範圍的平台，數據基礎設施、軟體平台...，幫助多個部門展開業務。
  * CEO應該提供啟動資金，再看能為公司帶來什麼效益。
3. 提供廣泛的AI培訓
  * 不只有工程師需要，而在公司中每一個層級的人都需要。
4. 發展AI戰略  
  在了解人工智能如何應用於公司後再制定策略比較好，避免在第一步就制定AI策略，不然可能會採取過於學術化的策略。
  * 利用人工智慧創造針對你所在產業的優勢
  * 符合「人工智慧良性循環」的設計策略
  * 考慮制定數據策略
    * 獲取戰略數據  
      例如，提供免費的照片共享服務、電子郵件服務、等等其他無法營利服務...去幫助以各種方式去蒐集數據。
    * 統一資料倉儲
  * 創造網路效應與平台優勢
    * 在「贏家通吃」的產業中，人工智慧可以成為加速器
5. 發展內部和外部溝通  
  與具有相關利益者進行溝通。
  * 投資人關係
  * 政府關係
  * 客戶/使用者教育

### A realistic view of AI  
我們需要對AI有符合現實的看法，不要過於樂觀，也不要過於悲觀。
* Too optimistic:
  * AI可以解決人類的所有問題，並創造烏托邦。
* Too pessimistic:
  * AI可能會有感知，並且殺死全人類。
* Just right:
  * AI是一個強大的工具，但同時他仍具有許多無法達成的事情。
  * 他有一些潛在的危害，ex.偏見、不公平、不準確的產出。

### Discrimination / Bias  
微軟的研究小組發現AI可以學習刻板印象。  
透過閱讀網路上的文本AI可以學習單詞，並進行推理類比。  
ex.  
>問AI:男人跟女人的對應關係相當於父親和誰?  
 AI會回答，母親。這反映了網路中單詞常用的方式。  
 同理，問AI:男人跟女人的對應關係相當於國王對什麼?  
 AI會回答，皇后。
>
>但當問AI::男人對工程師就像女人對什麼?  
AI會回答，女人對家庭主婦。  
>
>但正確沒有偏見的回答應該是女人對工程師。

**AI的偏見是如何產生的?**  
人工智能系統儲存單詞的方式是使用一組數字，ex.(1,1)代表man，這些數字座標代表了這個詞的典型用法，而computer programmer是(3,2)，woman是(2,3)。  
當問AI男人對工程師的關係時，AI會求出往右2，往上1的向量座標，而相對應的女人的對應詞，也會依據女人的座標再往右2，往上1，得出(4,4)，而(4,4)即為家庭主婦的座標。  
  

**去除AI偏見的方法**  
* Technical solutions:
  * 研究人員發現將具有偏見的數字零初始化後可有效減少偏見。
  * 盡量使用具有較少偏見的數據訓練。  
    ex.人臉辨識系統，應該包含多種種族。
* AI團隊要求系統有寮好透明度和審計流程，以及時發現AI中存在的偏見，並採取措施去糾正他。
* 擁有多元背景的員工，他們可以幫助數據更具多元化與包容性。

### Adversarial attacks on AI  
如果對圖片做出人眼無法發現的改變將導致AI辨識錯誤。
  
在計算機安全學中，對安全性的攻擊指讓系統規避他應該做的事情，同樣的一種針對AI的對抗攻擊是試圖讓他去做一些原本不打算做的事情，例如，愚弄他，以使他輸出不正確的分類。  
此外除了透過修改圖像外，也可以使用物理方式使AI辨識錯誤，例如被拍攝者配戴一副特殊眼鏡以使AI以為影像中的主角是一位著名女演員。  
或者在STOP停車標誌上塗鴉，也會使AI辨識錯誤，AI會不知道這是停車標誌。  

### Adversarial defenses  
* 新技術使AI較難被攻擊，但可能帶來一些代價，例如，速度變慢。
* 及時的來回交戰。

````python
# Example of global variable

artist = "Michael Jackson"
def printer1(artist):
    internal_var1 = artist
    print(artist, "is an artist")
    
printer1(artist)
# try runningthe following code
printer1(internal_var1) 
````

<b>We got a Name Error:  <code>name 'internal_var' is not defined</code>. Why?</b>

It's because all the variables we create in the function is a <b>local variable</b>, meaning that the variable assignment does not persist outside the function.

But there is a way to create <b>global variables</b> from within a function as follows:


````python
artist = "Michael Jackson"

def printer(artist):
    global internal_var 
    internal_var= "Whitney Houston"
    print(artist,"is an artist")

printer(artist) 
printer(internal_var)
````

## Errors vs. Exceptions
Hold on, what is the difference between errors and exceptions? Well, errors are usually big problems that come from the computer or the system. They often make the program stop working completely. On the other hand,` exception`s are more like issues we can control. They happen because of something we did in our code and can usually be fixed, so the program keeps going.

## Exception Handling
```try:... except:...```
一旦發生錯誤，啟動except搜索與錯誤匹配的異常，一旦找到正確的異常來處理錯誤，執行except內的敘述。  
```finally:```語句放置在except之下，為無論錯誤有無發生都一定要執行的語句。

| 方面 | 錯誤 | 異常 |
| :-- | :-- | :-- |
| 起源 | 錯誤通常是由環境、硬體或作業系統引起的。 | 異常通常是程式內部有問題的程式碼執行所導致的。 |
| 性質 | 錯誤通常很嚴重，會導致程式崩潰或異常終止。 | 異常一般不那麼嚴重，可以捕獲並處理，以防止程式終止。 |
| 處理 | 程式通常不會捕獲或處理錯誤。 | 異常可以使用 try-except 區塊進行捕獲並妥善處理，使程式繼續執行。 |
| 示例 | 示例包括由於語法錯誤導致的 “SyntaxError” 或變數未定義導致的 “NameError”。 | 示例包括除零時的 “ZeroDivisionError” 或嘗試開啟不存在的檔案時的 “FileNotFoundError”。 |
| 分類 | 錯誤不分類。 | 異常根據其性質分為各種類別，如 “ArithmeticError”、“IOError”、“ValueError”等。 |r”等。 |

````python
# using Try- except 
try:
    # Attempting to divide 10 by 0
    result = 10 / 0
except:
    # Handling the ZeroDivisionError and printing an error message
    print("Error: Cannot divide by zero")
finally:
    print("always running block")
# This line will be executed regardless of whether an exception occurred
print("outside of try and except block")
````

````python
a = 1

try:
    b = int(input("Please enter a number to divide a"))
    a = a/b
    print("Success a=",a)
except ZeroDivisionError:
    print("The number you provided cant divide 1 because it is 0")
except ValueError:
    print("You did not provide a number")
except:
    print("Something went wrong")
else: # code to execute if there is no exception
    print("NO error happen!")
finally:
    print("always running block")
````

## Common Exceptions in Python
* `ZeroDivisionError`:錯誤發生在試圖將一個數字除以零時。
* `ValueError`: 此錯誤發生在程式中使用了不適當的值。  
  例如，試圖將一個非數字字串轉換為整數時會出現此錯誤
* `FileNotFoundError`: 當試圖存取一個不存在的檔案時會遇到此異常。
* `IndexError`: 當使用一個超出有效索引範圍的索引來存取列表中的元素時。
* `KeyError`: 當試圖存取字典中不存在的鍵時。
* `TypeError`: 當一個物件以不相容的方式使用時會發生 TypeError。  
  例如，試圖將字串與整數連接在一起時。
* `AttributeError`: 當嘗試存取一個物件上不存在的屬性或方法時。
* `ImportError`: 當試圖匯入一個不存在的模組時會遇到此錯誤。

下一步是一個稱為建構函式（constructor）的特殊方法 `__init__`，用於初始化物件。輸入的是資料屬性。`self` 包含集合中的所有屬性。例如，`self.color` 會給出屬性 `color` 的值，而 `self.radius` 會給出物件的半徑。我們還有一個方法 `add_radius(r)`，這個方法將 `r` 的值加到屬性 `radius` 上。要存取半徑，我們使用語法 `self.radius`。

<img src="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/labs/Module%203/images/ClassesCircle.png" width="600">


````python
# Create a class Circle

class Circle(object):
    
    # Constructor
    def __init__(self, radius=3, color='blue'):
        self.radius = radius
        self.color = color 
    
    # Method
    def add_radius(self, r):
        self.radius = self.radius + r
        return(self.radius)
    
    # Method
    def drawCircle(self):
        plt.gca().add_patch(plt.Circle((0, 0), radius=self.radius, fc=self.color))
        plt.axis('scaled')
        plt.show() 
````

````python
# Create a new Rectangle class for creating a rectangle object

class Rectangle(object):
    
    # Constructor
    def __init__(self, width=2, height=3, color='r'):
        self.height = height 
        self.width = width
        self.color = color
    
    # Method
    def drawRectangle(self):
        plt.gca().add_patch(plt.Rectangle((0, 0), self.width, self.height ,fc=self.color))
        plt.axis('scaled')
        plt.show()
        
````

````python
# create an object of class circle
RedCircle = Circle(10, 'red') # name of object = constructor(parameters)
````

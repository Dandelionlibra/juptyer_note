## What is machine learning?  
定義："Field of study that gives computers the ability to learn without being explicitly programmed." -- Arthur Samuel  

通常給予的學習機會越多，AI表現的會越好。  

Machine learning algorithms  
* Supervised learning  
  現代應用最多的機器學習演算法，其進步和創新速度最快。
* Unsupervised learning
* Recommender systems
* Reinforcement learning  

至今最常使用的演算法類型是監督學習、無監督學習、推薦系統。

### Supervised learning:
* **Regression algorithms**:  
是指學習input(x)映射到output(y)的算法。  
監督學習的關鍵特徵是需要給出學習事例(提供給定x的輸出y)，透過train set，機器學會在沒有提供輸出y的前提下，就可以藉由x預測、計算出準確的y。  
ex.預測的線段可以是一條直線，或者一個函數，使預測的結果更符合事實。  
這種房價預測稱為回歸(Regression)，回歸是試圖在無限多可能的數字中預測一個數字，如下圖中的房價。  
  

* **Classification algorithms**:  
機器學習系統試圖將輸入分類，
ex.偵測腫瘤，機器學習系統需要將腫瘤分成良性(0)或惡性(1)。  
分類與回歸不同，分類僅只有有限的輸出，分類演算法可以預測限定好的類別，而不包含介於兩者之間的可能性。  

  

**supervised learning**:  
* 從正確答案(標有標籤)中學習。
* 主要分為兩種類型：Regression、Classification，依照需求去做選擇。

Notation:  
Input(x): $x$，稱為輸入變量("input" variable feature)，ex.x=2104  
Output(y): $y$，稱為輸出變量、目標變量("output" variable、"target" variable)ex.y=400  
每行表不同的訓練示例，以$m$表示訓練示例的總數，上方表格中的$m=47$。  
$(x,y)$表示單個訓練示例，ex.$(x,y)=(2104,400)$。  
$(x^{(i)},y^{(i)})$表示第$i^{th}$個訓練示例，$i$是訓練集的索引，ex.$(x^{(1)},y^{(1)})=(2104,400)$。


監督學習中的訓練集包含輸入變量與輸出變量，接著監督學習演算法會產生函數$f$，$f$的功能是取一個新的輸入$x$，接著輸出他的預測值$\hat{y}$，在機器學習中，$\hat{y}$是$y$的估計值或預測。  
函數$f$稱為模型(model)，輸入$x$稱為feature，輸出$y$稱為prediction，模型的預測是y的估計值，當標示僅為$y$時代表他是輸入的target值，即訓練集中的真實值，而$\hat{y}$是一個估計值，他可能是也可能不是實際的真實值。  
  


**該如何表示函數f?**  
假設將f當作一直線，可標示成$f_{w,b}(x)=wx+b$，$f$是一個以$x$做為輸入的函數，$w,b$為這個模型的參數，根據w、b的值，$f$將輸出預測$\hat{y}$的值，可簡寫成$f(x)$就好。  
而在機器學習中，參數是可以在訓練期間調整，以去改進模型的變量，有時候參數也被稱為係數(coefficients)或權重(weight)。  
$f_{w,b}(x)=wx+b$為帶有一個變量的線性回歸，其只有一個$x$作為輸入變量，這種特性線性模型的另種名稱為單變量線性回歸(Univariate linear regression)。  


**成本函數(cost function)**  
成本函數將告訴我們模型的表現。  
ex. 假設有如下表的一組training set:  

| size in feet$^2$ | price in $1000's |
| :-- | :-- |
|(1) 2104 | 400 |
|(2) 1416 | 232 |
|(3) 1534 | 315 |
|(4) 852 | 178 |
| ... | ... |
|(47) 3210 | 870 |  

用模型$f_{w,b}(x)=wx+b$去擬合這個訓練集，當為$w、b$選擇不同的值，會得到不同的函數$f(x)$。  
下方第一張圖中$f(x)=0*x+1.5$，因此$f(x)$始終是常量值，他總是預測$y$的估計值為1.5，即$\hat{y}$總是等於$b$，y945; 這張圖的$b$也被稱為$y$的截距(intercept)。  

第二張圖中$f(x)=1.5x$，斜率是0.5/1，$w$為直線的斜率(=0.5)。  

第三張圖中$f(x)=1.5x+1$，$b$為$y$的截距(=1)，$w$為直線的斜率(=0.5)。  


將每個點的預測值$\hat{y}$減去$y$後平方的數據加總，以上方的表格為例，即加總$m=47$次，且為了避免因為數據量增加($m$增加)而得到很大或很小的成本函數($J(w,b)$)，所以還需要對總平方誤差和求平均。  
但實際上，機器學習中使用的成本函數實際上在對總平方誤差和求平均時還需要除以2，目的是為了讓之後的計算看起來更加簡潔，但無論有沒有除2，都是有效的成本函數表示。  
因為需要平方，因此也稱平方誤差成本函數(squared error cost function)。  
  

還存在其他成本函數的計算方式，但對於線性回歸，這種成本函數至今為止最常用。

* model:$f_{w,b}(x)=wx+b$
* parameters:$w,b$
* cost function:$J(w,b)=\frac{1}{2m}\sum\limits_{i=0}^{m}(f_{w,b}(x^{(i)}-y^{(i)})^2$

* goal:$\mathop{minimize}\limits_{w,b}J(w,b)$，最小化$J(w,b)$

### Unupervised learning:  
給予進行無監督學習的數據僅帶有輸入x，而不會帶標籤有正確輸出的y，而因為沒有提供正確輸出，因此演算法需要自己找出數據中的分類。
繼續以腫瘤診斷輸入為例，他沒有被要求診斷出腫瘤類型，因為沒有提供給他標籤，而進行無監督學習的目標是尋找數據中的結構、某種模式或一些有用的東西。  

* **Clustering algorithms**:Group similar data pointers togeter.  
  他會將未標記的數據放入不同的集群中。ex.Google news使用類聚演算法，將相關新聞分類，因為每天的報導太多了，很難透過人工逐一分類，。
  類聚演算法找到提及相似詞語的文章並將他們分群，這種演算法可以自己找出哪些詞暗示是有意義的並分群。
  也可以為顧客分群。
    
  另種例子是DNA數據，給予DNA微陣列數據的圖片，每列都是不同的個體，而每行控制不同個體相同基因片段。在沒有告訴他有什麼類型DNA的前提下，讓群聚演算法自動尋找相似並分類不同類型的人。
    

* **Anomaly detection(異常檢測)**:檢測異常事件。  
  ex.金融體系中的詐欺檢測，可以檢測是否存在可能是詐欺的異常交易等...  

* **Dimensionality reduction(降維)**:可以將一個大數據集壓縮成一個小數據，並盡可能的減少丟失的訊息。

## 01 Introduction to Generative AI

* 甚麼是人工智慧?  
  AI對於電腦科學就像物理對於科學，AI涵蓋一系列理論與方法，目標是打造能模仿人類思考和行動的機器。
* 機器學習：  
  為AI的子領域，會根據輸入資料訓練模型的系統，訓練好的模型能根據全新或未知
  資料進行預測。
  > Machine Learning gives the computer the ability to learn without explicit programming
  * 監督式模型：監督式模型會使用「標籤」機制，會根據過去的樣本學習，並給出預測值。
  * 非監督式模型：無標籤資料，透過觀察資料，看看其中是否存在有一定關連性而形成的群組。
  * 強化學習
  * 深度學習： 採用人工類神經網路，有多層神經元，與機器學習相比，能處理更複雜的模式。  
    類神經網路可使用標籤(少量)和無標籤(大量)資料，稱為半監督式學習。標籤資料可以幫助學會目標工作的基本概念，無標籤資料則培養類神經網路面對新樣本的「一般化」能力。

* 生成式AI：  
  是深度學習的分支，為一種人工智慧技術，能生成多種類型的內容，包含文字、圖像、音訊、合成資料。  
  採用人工類神經網路 能透過監督式、非監督式和半監督式方法處理標籤資料和無標籤資料。
* 大型語言模型也是深度學習的分支。

**深度學習模型**  
通常也稱為機器學習模型。  
* 生成式(Generative)模型：從現有資料中學習機率分布，然後生成新的資料實例(新內容)。通常輸出:自然語言、圖像、影片...。
* 判別式(Discrimnative)模型：用來分類或預測資料點的標籤，會使用含有標籤的資料集，模型會判斷資料點特徵與標籤之間的關係，通常輸出:離散資料、數字、分類、可能性...。

**傳統機器學習的監督式學習**過程，會運用訓練程式碼和標籤資料來建構模型，該模型能預測結果，也可以將目標物分類或分群。  


**生成式AI** 能使用訓練程式碼和所有類型的標籤/無標籤資料來建構基礎模型，例如文字、程式碼、圖片、音訊、影片...

* PaLM (Pathways 語言模型)
* LaMDA (對話應用程式語言模型)  
都會從多種網路來源擷取大量資料並建構出基礎語言模型。

生成式AI能夠透過現有內容學習，並根據所學生成新內容。從現有內容中學習的過程稱為「訓練」訓練完畢後，就會產生一個統計模型，只要給予提示，生成式AI學習資料的基礎架構後，會運用統計模型預測預期的回覆，生成新內容，內容近似於先前訓練階段時所用的資料。  

大型語言模型是生成式AI的一種，透過接收自然語言提示產出新的文字組合。  
生成式AI最主要的厲害之處在於Transformer模型，它由編碼器和解碼器組成，編碼器會將輸入序列編碼，然後傳給解碼器，解碼器會針對相關工作學習如何解讀其表徵，不過有時Transformer會衍生一些「幻覺」，幻覺是模型生成的字詞或詞組，這類內容通常邏輯不通或有文法錯誤。
* 模型訓練的資料不足
* 訓練資料品質不佳
* 給模型的背景知識不足
* 未設下足夠的限制  
> Hallucinations make the output text difficult to understand and make the model more likely to generate incorrect or misleading information.

「提示(Prompting)」是一小段文字，會以輸入內容的形式提供給大型語言模型(簡稱 LLM)，可以控制模型的輸出內容，「提示設計」是指找出合適提示的過程，目的是讓LLM生成理想的輸出結果，生成式AI的回覆絕大部分取決於提供的訓練資料，它會分析輸入資料的模式和結構並從中學習。

**基礎模型**  
規模比先前提到的都還要大，這類大型AI模型會以巨量資料預先訓練，並能加以調整或微調，以便處理各種工作，ex.情緒分析、圖像說明生成、物件辨識...

## 02 Introduction to Large Language Models

大型語言模型(LLM)是深度學習的分支，LLM與生成式AI都屬於深度學習，且有部分領域重疊。  
> Large Language Models refer to large, general-purpose language models that can be pre-trained and then fine-tuned for specific purposes

大型語言模型被用來解決常見的語言問題，ex.文字分類、回答問題、文件摘要、生成文字。且這類模型還可以用規模較小的產業資料集做進一步調整，以解決各領域的問題。

**LLM定義**
* Large:  
  * 大型訓練集
  * 眾多參數數量
* General purpose:  
  指模型足以解決常見問題。
* Pre-trained and fine-tuned:
  使用大型訓練集針對一般用途預先訓練的大型語言模型，再用較小型的資料集微調，最終產生具有特定功能的模型。

**LLM用途**  
* 透過單個模型處理多項工作
* 調整大型語言模型去解決特定問題時，需要的訓練資料不多(少樣本甚至零樣本)
  * 少樣本：利用極少量資料訓練模型
  * 零樣本：可以辨識先前未在訓練中明確學到的事物
* 隨著資料量、參數量增加，大型語言模型的成效也會持續增長。

**例子**  
* Pathways Language Model(PaLM)
* Language Model for Dialog Applications(LaMDA)
* 為Transformer模型，由編譯器與解碼器構成。

傳統機器學習開發作業需要專業能力、訓練樣本、運算時間及硬體，相較於LLM開發作業需具備的條件更多。

* 提示設計(Prompt design)：  
  指找出合適提示的過程，根據使用者要求系統執行的特定工作來建立相應提示。
* 提示工程(Prompt engineering)：  
  指為提升成效而建立相應提示的過程，過程中可能需要運用特定領域知識以提供理想輸出結果範例，或使用能在特定系統中發揮良好成效的關鍵字。

**提示設計**不可或缺，**提示工程**則只有在系統需達到高準確率或高效能時才需進行。

LLM分為三種，各種模型的提示方法都不同。：
* 通用語言模型(Generic language models)：Predict the next word(technically, token) based on the language in the training data.  
  
* 指令調整模型(Instruction tuned)：根據輸入中的指令預測回復。
Trained to predict a response to the instructions given in the input.

* 對話調整模型(Dialog tuned)：Train to have a dialog by predicting the next response.  
  對話調整模型是特殊的指令調整模型，使用者通常會以問句向聊天機器人提出要求，這類模型適合用在一來一往的長篇對話中。  
  模型具有「關聯思考推論」能力，指的是模型輸出文字時，如果要它們先解釋背後的推理過程 它們更能提供正確的解答。

## 03 Introduction to Responsible AI

為了開發負責任的AI技術，必須瞭解可能出現的問題、限制、非預期的結果，科技會反映社會現況，如果沒有適當的做法AI可能會帶來相同的問題或偏見，甚至使情況加劇。

Responsible AI其實並沒有通用定義或簡潔的檢查清單，因此，許多機構會擬定自己的AI開發原則。  

以google為例：
1. AI 應對社會有益，
無論任何案 都必須考量多項社會和經濟
2. AI 應避免塑造或加深不公平的偏誤
3. AI 應設有防護措施並通過安全測試
4. AI 應對使用者負責，例如，提供適當機會讓人們提出意見、相關說明和訴求
5. AI 應採用隱私設計原則
6. AI 應堅持以高標準追求科學卓越
7. AI 應以上述原則做為應用基準因素


## 04 Prompt Design in Vertex AI

生成式AI模型生成内容時需要具备以下功能：  

* 了解如何執行新任務：如果希望模型執行特定任務，需能够自定義模型。
* 訪問外部信息：除了訓練數據之外，還需要能夠訪問訓練數據之外的訊息。
* 過濾有害內容：模型可能生成令人不適的內容，因此需要安全過濾器屏蔽有害的提示或回答。

## 05 Responsible AI: Applying AI Principles with Google Cloud

大型語言模型疑慮：
* 幻覺：產生的內容不切實際、虛構、完全捏造
* 事實性：與生成式AI所生資訊的準確性、真實性有關
* 擬人化：非人類實體具有類似人類的特質、特徵、行為

原因：  
* 缺乏專屬資源(資金、人員、技術)打造符合倫理的AI系統
* 缺乏種族、性別、地理區域多樣化的團隊共同開發AI系統
* 倫理方面缺乏AI行為準則、評估偏差的能力

定義AI開發原則後，建立AI控管機制的下一步是制定審查程序並付諸實行，但將AI開發原則做為指引，並不能立即解決所有與AI道德倫理相關的疑慮。  

穩健的AI控管程序是建立目標的關鍵第一步，負責任的AI技術工具只能在責任目標明確時提供幫助。  

誤解：
* 聘僱有道德的人員就能保證AI產品符合道德倫理?  
  兩個有強烈道德感的人在評估相同情況或AI解決方案時，會根據各自的經驗和背景得出完全不一樣的結論。
* 為負責任的AI技術建立檢查清單?  
  每項產品都有獨一無二的技術細節和使用情境，因此需要專屬的評估，若只依照檢查清單行事，可能會限制批判性思考並造成道德盲點


Google建立了正式的審查委員會架構去評估專案、產品、交易是否符合他們定義的AI開發原則。委員會架構其中的負責任創新團隊，會指引實行AI開發原則審查的不同產品團隊、為AI開發原則訂定常用解釋、確保公司決策的一致性。  
負責任創新團隊中包含：
* 使用者研究人員
* 社會科學家
* 倫理學家
* 人權專家
* 政策與隱私權顧問
* 法律專家
* ...帶來多元觀點與專業知識

委員會中第二個團隊是由各專業領域的資深專家所組成的，負責提供技術、功能和應用程式方面的專業知識，會提供與新興技術和主題相關的策略和指南，並在有需要時提供審查諮詢。

第三個AI控管團隊是由高階主管組成的委員會，這些主管負責處理最複雜且最困難的問題，包括影響多種產品和技術的決策，他們身為接受提報的團體會做出開先例的複雜決策，並向公司最高層級負責。

最後是在特定產品領域中既有的專屬AI控管與審查委員會，他們會與負責任創新團隊密切合作，並在技術、應用情況、訓練資料、社會背景...
以及如將 I 整合至實際工作環境等各面 考量該產品領域的特殊。

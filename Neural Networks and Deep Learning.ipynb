## What is a Neural Network?

增加特徵，第一層是輸入層，第二層為全連接層(每個輸入特徵都會連接到層的每個神經元)，最後一層為輸出層。  
  
神經網路只要給定足夠多的訓練樣本，就可以很好的擬和出一個函數建立(x,y)間的映射關係。

### Logistic regression algorithm  
是一種二元分類問題。



輸入64*64大小的圖片，圖片依照RGB建立三個實數矩陣。  
輸入特徵x的維度：$n_x=64*64*3=12288$  

  


**Notation**  
$(x,y)=\left\{
\begin{aligned}
x\in R^{n_x} \\
y\in\{0,1\}
\end{aligned}
\right.$  

$m$個訓練樣本:=$(x^{(1)},y^{(1)})、(x^{(2)},y^{(2)})...(x^{(m)},y^{(m)})$  
$m_{train}、m_{test}$區分訓練樣本。  

$X=\vec{x}=\left[
\begin{array}{ccc}
    ⋮ \\
    x^{(1)} \ x^{(2)} \ ... \ x^{(m)}\\
    ⋮ 
\end{array}
\right]$  高度是維度$n$，寬度是樣本數$m$。  
$X=R^{n_x*m}$  
<font color="red">※X.shape=$(n_x,m)$</font>  

$Y=\left[
\begin{array}{ccc}
    y^{(1)} \ y^{(2)} \ ... \ y^{(m)}
\end{array}
\right]$  
$Y=R^{1*m}$  
<font color="red">※Y.shape=$(1,m)$</font>  

給予$\{(x^{(1)}, y^{(1)}),...,(x^{(m)}, y^{(m)})\}$，希望$\hat{y}^{(i)}\approx y^{(i)}$  
$\hat{y}=P(y=1|x)\rightarrow$輸入$x$後$y$等於$1$的機會， $0\leq\hat{y}\leq 1$  
$x=R^{n_x}$  
參數：$w\in R^{n_x},b\in R$  
輸出：$\hat{y}=\sigma(w^Tx+b)$, 使用$\sigma$使$\hat{y}$的範圍在0到1之間。  
$\sigma(z)=\frac{1}{1+e^{-z}}$，若z很大，$\sigma(z)\approx\frac{1}{1+0}=1$；若z很小，$\sigma(z)\approx\frac{1}{1+\propto}=0$


#### cost function  
Loss(error) function: $L(\hat{y},y)=\frac{1}{2}(\hat{y}-y)^2\rightarrow-(ylog\hat{y}+(1-y)log(1-\hat{y}))$  
第一個式子因為會產生多個局部最小值，因此修改成後面ㄉ式子，使其只有一個最小值。

假設$y=1$，$L(\hat{y},y)=-log\hat{y}$，希望$log\hat{y}$大$\rightarrow\hat{y}$大  
假設$y=0$，$L(\hat{y},y)=-log(1-\hat{y})$，希望$log(1-\hat{y})$大$\rightarrow\hat{y}$小  
Cost function: $J(w,b)=\frac{1}{m}\sum\limits_{i=1}^{m}L(\hat{y}^{(i)},y^{(i)})=-\frac{1}{m}\sum\limits_{i=1}^{m}[y^{(i)}log\hat{y}^{(i)}+(1-y^{(i)})log(1-\hat{y}^{(i)})]$

#### Gradient Descent  

$J$代價函數是一個凸函數(convex function)，尋找使$J(w,b)$最小的$w,b$。  
$w$可以是更高的維度。  
梯度下降法以隨機的初始位置開始朝向最陡的下坡方向走一步。  


repeat{  
　$w_j=w_j-{\alpha}\textcolor{magenta}{\frac{∂}{∂w_j}J(\vec{w},b)}$  
　$b=b-{\alpha}\textcolor{blue}{\frac{∂}{∂b}J(\vec{w},b)}$  
}  

##### **Derivatives**

$f(a)=a^2\rightarrow\frac{d}{da}f(a)=2a$  
ex.  
a=2, f(a)=4  
a=2.001, f(a)=4.004001  

$f(a)=ln(a)=log_e(a)\rightarrow\frac{d}{da}f(a)=\frac{1}{a}$  

**Computation Graph**  

$J(a,b,c)=3(a+bc)=
\left\{
\begin{aligned}
u=bc \\
v=a+u \\
J=3v \\
\end{aligned}
\right.$  
  
  
令$v=11\rightarrow 11.001$  
則$J=33\rightarrow 33.003$  
計算J對於v的導數：$J=3v\rightarrow\frac{dJ}{dv}=3;f(a)=3a\rightarrow\frac{df(a)}{da}=3$  
v增量了3倍變為J。


**Gradient Descent**  
$z=w^Tx+b$  
$\hat{y}=a=\sigma(z)$  
$L(a,y)=-(ylog(a)+(1-y)log(1-a))$  

**反向計算導數**  
參數:$x_1,w_1,x_2,w_2,b\textcolor{red}{\rightarrow} z=x_1w_1+x_2w_2+b\textcolor{red}{\rightarrow}\hat{y}=a=\sigma(z)\textcolor{red}{\rightarrow} L(a,y)$  

計算損失函數的導數$\downarrow$  
$更新=\left\{
\begin{aligned}
w_1=x_1-\alpha dw_1 \\
w_2=x_2-\alpha dw_2 \\
b=b-\alpha db \\
\end{aligned}
\right.
\textcolor{red}{\leftarrow}
\frac{∂L}{∂w_1}=
\left\{
\begin{aligned}
dw_1=x_1dz \\
dw_2=x_2dz \\
db=dz \\
\end{aligned}
\right.
\textcolor{red}{\leftarrow}
dz=\frac{dL(a,y)}{dz}=a-y(=\frac{dL}{da}\frac{da}{dz}=(-\frac{y}{a}+\frac{1-y}{1-a})*(a(1-a)))
\textcolor{red}{\leftarrow} da=\frac{dL(a,y)}{da}=-\frac{y}{a}+\frac{1-y}{1-a}$  

https://community.deeplearning.ai/t/derivation-of-dl-dz/165

### A Note on Python/Numpy Vectors

```python
import numpy as np
ran = np.random.randn(5)
ran
```

```python
print(ran.shape)
```

```python
# ran的轉置
print(ran.T)
```

```python
print(np.dot(ran,ran.T))
```

上方ran為一維的數組(一個中括號)，因為在計算時會發生不是我們所需要的結果，因此需將其改為下方ran2的二維數組(兩個中括號)

```python
# 使ran2成為5*1的列向量
ran2 = np.random.randn(5,1)
ran2
```

```python
print(ran2.shape)
```

```python
# ran2的轉置
print(ran2.T)
```

```python
print(np.dot(ran2,ran2.T))
```

```python
# 確認ran2是否是column vector
# false時觸發錯誤
assert(ran2.shape==(5,1))
assert(ran2.shape==(1,5))
```

```python
# 改變形狀
ran2 = ran2.reshape(1,5)
assert(ran2.shape==(5,1))
assert(ran2.shape==(1,5))
```

## What is a Neural Network?

增加特徵，第一層是輸入層，第二層為全連接層(每個輸入特徵都會連接到層的每個神經元)，最後一層為輸出層。  
  
神經網路只要給定足夠多的訓練樣本，就可以很好的擬和出一個函數建立(x,y)間的映射關係。

### Supervised Learning with Neural Networks

輸入x，並通過x學習到能將x映射到輸出y的函式。  
* Real Estate$\rightarrow$相對標準的神經網路(NN)
* Online Advertising$\rightarrow$相對標準的神經網路(NN)
* Photo tagging$\rightarrow$卷積神經網路(CNN)
* Speech recognition$\rightarrow$由時序組成的數據，循環神經網路(RNN)
* Machine translation$\rightarrow$循環神經網路(RNN)
* Autonomous driving$\rightarrow$混和多種神經網路結構，因為用於多種應用  



sigmoid函數在梯度幾乎為0時，學習的速度非常慢，ReLU（rectified linear unit）對於所有的正輸入梯度都是1，不會有慢慢變為0的過程，因此速度提高很多。

## 01 Introduction to Generative AI

* 甚麼是人工智慧?  
  AI對於電腦科學就像物理對於科學，AI涵蓋一系列理論與方法，目標是打造能模仿人類思考和行動的機器。
* 機器學習：  
  為AI的子領域，會根據輸入資料訓練模型的系統，訓練好的模型能根據全新或未知
  資料進行預測。
  > Machine Learning gives the computer the ability to learn without explicit programming
  * 監督式模型：監督式模型會使用「標籤」機制，會根據過去的樣本學習，並給出預測值。
  * 非監督式模型：無標籤資料，透過觀察資料，看看其中是否存在有一定關連性而形成的群組。
  * 強化學習
  * 深度學習： 採用人工類神經網路，有多層神經元，與機器學習相比，能處理更複雜的模式。  
    類神經網路可使用標籤(少量)和無標籤(大量)資料，稱為半監督式學習。標籤資料可以幫助學會目標工作的基本概念，無標籤資料則培養類神經網路面對新樣本的「一般化」能力。

* 生成式AI：  
  是深度學習的分支，為一種人工智慧技術，能生成多種類型的內容，包含文字、圖像、音訊、合成資料。  
  採用人工類神經網路 能透過監督式、非監督式和半監督式方法處理標籤資料和無標籤資料。
* 大型語言模型也是深度學習的分支。

**深度學習模型**  
通常也稱為機器學習模型。  
* 生成式(Generative)模型：從現有資料中學習機率分布，然後生成新的資料實例(新內容)。通常輸出:自然語言、圖像、影片...。
* 判別式(Discrimnative)模型：用來分類或預測資料點的標籤，會使用含有標籤的資料集，模型會判斷資料點特徵與標籤之間的關係，通常輸出:離散資料、數字、分類、可能性...。

**傳統機器學習的監督式學習**過程，會運用訓練程式碼和標籤資料來建構模型，該模型能預測結果，也可以將目標物分類或分群。  


**生成式AI** 能使用訓練程式碼和所有類型的標籤/無標籤資料來建構基礎模型，例如文字、程式碼、圖片、音訊、影片...

* PaLM (Pathways 語言模型)
* LaMDA (對話應用程式語言模型)  
都會從多種網路來源擷取大量資料並建構出基礎語言模型。

生成式AI能夠透過現有內容學習，並根據所學生成新內容。從現有內容中學習的過程稱為「訓練」訓練完畢後，就會產生一個統計模型，只要給予提示，生成式AI學習資料的基礎架構後，會運用統計模型預測預期的回覆，生成新內容，內容近似於先前訓練階段時所用的資料。  

大型語言模型是生成式AI的一種，透過接收自然語言提示產出新的文字組合。  
生成式AI最主要的厲害之處在於Transformer模型，它由編碼器和解碼器組成，編碼器會將輸入序列編碼，然後傳給解碼器，解碼器會針對相關工作學習如何解讀其表徵，不過有時Transformer會衍生一些「幻覺」，幻覺是模型生成的字詞或詞組，這類內容通常邏輯不通或有文法錯誤。
* 模型訓練的資料不足
* 訓練資料品質不佳
* 給模型的背景知識不足
* 未設下足夠的限制  
> Hallucinations make the output text difficult to understand and make the model more likely to generate incorrect or misleading information.

「提示(Prompting)」是一小段文字，會以輸入內容的形式提供給大型語言模型(簡稱 LLM)，可以控制模型的輸出內容，「提示設計」是指找出合適提示的過程，目的是讓LLM生成理想的輸出結果，生成式AI的回覆絕大部分取決於提供的訓練資料，它會分析輸入資料的模式和結構並從中學習。

**基礎模型**  
規模比先前提到的都還要大，這類大型AI模型會以巨量資料預先訓練，並能加以調整或微調，以便處理各種工作，ex.情緒分析、圖像說明生成、物件辨識...

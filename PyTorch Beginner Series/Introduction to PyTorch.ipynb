````python
# 初始化隨機張量
x = torch.rand(1, 10)       # 輸入向量 (1 x 10)
prev_h = torch.rand(1, 20)  # 前一個隱藏狀態 (1 x 20)
W_h = torch.rand(20, 20)    # 隱藏狀態的權重矩陣 (20 x 20)
W_x = torch.rand(20, 10)    # 輸入向量的權重矩陣 (20 x 10)

# 計算下一個隱藏狀態
# x.t() 是 PyTorch 中用來對張量 x 進行轉置操作的方法，轉置後的 x 的形狀變為 (10 x 1)，以便與 W_x (20, 10)相乘
i2h = torch.mm(W_x, x.t())  # 輸入到隱藏狀態的計算 (20 x 1)
h2h = torch.mm(W_h, prev_h.t())  # 前一個隱藏狀態到隱藏狀態的計算 (20 x 1)
next_h = i2h + h2h  # 將輸入和前一個隱藏狀態的貢獻相加
next_h = next_h.tanh()  # 應用雙曲正切函數作為非線性激活函數

# 計算損失 (例如，隱藏狀態所有元素的和)
loss = next_h.sum()

````

### Buliding module in pytorch

````python
import torch # 引入 PyTorch
import torch.nn as nn # 引入神經網絡庫, module 用於構建神經網絡層
import torch.nn.functional as F # 引入函數庫
````

![Buliding module in pytorch1.png](<attachment:Buliding module in pytorch1.png>)

````python
# 定義一個繼承自 nn.Module 的類，並實現其中的 forward 方法，這個類將作為我們的神經網絡模型。
class LeNet(nn.Module):
    # 在 __init__ 方法中，定義了網絡的結構，包括兩個卷積層以及三個全連接層。
    def __init__(self):
        super(LeNet, self).__init__()
        self.conv1 = nn.Conv2d(1, 6, 3) # 第一層卷積層，輸入通道數 1，輸出通道數 6，卷積核大小 5
        self.conv2 = nn.Conv2d(6, 16, 3) # 第二層卷積層，輸入通道數 6，輸出通道數 16，卷積核大小 5
        # an affine operation: y = Wx + b
        self.fc1 = nn.Linear(16 * 6 * 6, 120) # 第一層全連接層，輸入維度 16 * 6 * 6，輸出維度 120
        self.fc2 = nn.Linear(120, 84)
        self.fc3 = nn.Linear(84, 10)

    # forward 方法定義了數據如何在網絡中傳播。每一層都應用一個激活函數，然後進行最大池化，最後經過全連接層得到輸出。
    def forward(self, x):
        # Max pooling over a (2, 2) window
        x = F.max_pool2d(F.relu(self.conv1(x)), (2, 2))
        # If the size is a square you can only specify a single number
        x = F.max_pool2d(F.relu(self.conv2(x)), 2)
        # 進行形狀變換，將卷積層的輸出平整為一維張量，用於全連接層的輸入，-1 表示自適應
        x = x.view(-1, self.num_flat_features(x))
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        return x
    
    # num_flat_features 方法用於計算張量 x 的總特徵數(展平特徵的數量，用於在 forward 方法中將張量展平)，即除了 batch 維度之外的所有維度的乘積。
    def num_flat_features(self, x):
        size = x.size()[1:] # all dimensions except the batch dimension
        num_features = 1
        for s in size:
            num_features *= s
        return num_features
````

![Buliding module in pytorch2.png](<attachment:Buliding module in pytorch2.png>)

````python
net = LeNet()
print(net) # 輸出網絡結構

# batch_size x channels x height x width
input = torch.randn(1, 1, 32, 32) # 生成一個隨機輸入
print('\nImage batch shape:')
print(input.shape)

output = net(input) # 將輸入輸入到網絡中
print('\nRaw output:')
print(output)
# 
print(output.shape)
````

### Datasets and dataloaders

````python
# 這是一個 Jupyter Notebook 的魔法命令，用來啟用內嵌的圖形顯示模式。
# 它使得 matplotlib 生成的圖形直接顯示在 Notebook 內部，而不是彈出一個新的視窗。
%matplotlib inline

import torch
# PyTorch 的計算機視覺工具包，包含了常用的影像數據集、模型和影像處理變換。
import torchvision
import torchvision.transforms as transforms
````

![Datasets and dataloaders 1.png](<attachment:Datasets and dataloaders 1.png>)

`transforms.ToTensor()` 將 PIL 圖片或 NumPy 陣列轉換為 PyTorch 張量。影像的像素值會被縮放到 [0, 1] 的範圍內，且張量的形狀會被轉換為 (通道數, 高度, 寬度)

`Normalize` 是用來對影像的像素值進行正規化的操作。這裡的 (0.5, 0.5, 0.5) 是每個通道的平均值，(0.5, 0.5, 0.5) 是每個通道的標準差。  
正規化公式是 (x - mean) / std，其中 x 是像素值。  
這個例子中，這意味著輸入的像素值範圍會從 [0, 1] 轉換到 [-1, 1]

````python
transform = transforms.Compose(
    [transforms.ToTensor(),
     transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))])
````

![Datasets and dataloaders 2.png](<attachment:Datasets and dataloaders 2.png>)

加載了 CIFAR-10 訓練數據集，並應用了之前定義的數據轉換。  

`torchvision.datasets.CIFAR10`: PyTorch 中的一個內建數據集類別，用於加載 CIFAR-10 數據集。  
`root='./data'`:`root` 指定了數據集的存儲位置。這裡設置為 `./data`，表示數據集會被下載並存儲在當前目錄下的 `data` 資料夾中。  
`train=True`: 會加載 CIFAR-10 的訓練數據集。CIFAR-10 數據集分為訓練集和測試集，訓練集包含 50000 張影像，而測試集包含 10000 張影像。  
`download=True`: 如果數據集沒有被提前下載，這個參數會自動下載數據集。如果數據集已經存在於指定的 `root` 目錄下，它將不會再次下載。  
`transform=transform`: 這裡指定了我們之前定義的數據轉換（transform），這些轉換會在每次加載影像時自動應用。  

````python
trainset = torchvision.datasets.CIFAR10(root='./data', train=True,
                                        download=True, transform=transform)
````

![Datasets and dataloaders 3.png](<attachment:Datasets and dataloaders 3.png>)

`batch_size=4`:
每個批次的大小設定為 4。這表示 DataLoader 每次將從訓練數據集中抽取 4 張影像組成一個小批次，然後提供給模型進行訓練。較小的批次可以使模型更快地更新參數，但也可能導致噪聲較大。批次大小應根據硬件資源和具體任務進行調整。  
`shuffle=True`:
將 shuffle=True 表示在每個 epoch 開始時，會將訓練數據集的順序打亂。這有助於打破數據集中的順序依賴，並且可以提高模型的泛化能力。  
`num_workers=2`:
這個參數設置了數據加載時使用的工作線程數量。num_workers=2 表示使用 2 個子進程來加載數據。這可以加速數據的加載和預處理過程。具體的線程數應根據硬件性能來調整。

````python
trainloader = torch.utils.data.DataLoader(trainset, batch_size=4,
                                            shuffle=True, num_workers=2)
````

### Training your python model

````python
%matplotlib inline

import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim

import torchvision
import torchvision.transforms as transforms

import matplotlib
import matplotlib.pyplot as plt
import numpy as np
````

training and test datasets. Run the cell below to make sure the dataset is downloaded. 

````python
transform = transforms.Compose(
    [transforms.ToTensor(),
     transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))])

trainset = torchvision.datasets.CIFAR10(root='./data', train=True,
                                        download=True, transform=transform)
trainloader = torch.utils.data.DataLoader(trainset, batch_size=4,
                                            shuffle=True, num_workers=2)
testset = torchvision.datasets.CIFAR10(root='./data', train=False,
                                        download=True, transform=transform)
testloader = torch.utils.data.DataLoader(testset, batch_size=4,
                                            shuffle=False, num_workers=2)
classes = ('plane', 'car', 'bird', 'cat',
            'deer', 'dog', 'frog', 'horse', 'ship', 'truck')
````

Run our check on the output from DataLoader:

````python
import matplotlib.pyplot as plt
import numpy as np

# functions to show an image
def imshow(img):
    img = img / 2 + 0.5     # unnormalize
    npimg = img.numpy()
    plt.imshow(np.transpose(npimg, (1, 2, 0)))
    # plt.show()

# get some random training images
dataiter = iter(trainloader)
images, labels = next(dataiter)

# show images
imshow(torchvision.utils.make_grid(images))
# print labels
print(' '.join('%5s' % classes[labels[j]] for j in range(4)))
````

這是要訓練的模型，是之前討論過的LeNet的變體。

````python
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.conv1 = nn.Conv2d(3, 6, 5)
        self.pool = nn.MaxPool2d(2, 2)
        self.conv2 = nn.Conv2d(6, 16, 5)
        self.fc1 = nn.Linear(16 * 5 * 5, 120)
        self.fc2 = nn.Linear(120, 84)
        self.fc3 = nn.Linear(84, 10)

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = x.view(-1, 16 * 5 * 5)
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        return x

net = Net()
````

````python
criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(net.parameters(), lr=0.001, momentum=0.9)
````

![Training your python model 1.png](<attachment:Training your python model 1.png>)

````python
for epoch in range(2):  # loop over the dataset multiple times
    
        running_loss = 0.0
        for i, data in enumerate(trainloader, 0):
            # get the inputs; data is a list of [inputs, labels]
            inputs, labels = data
    
            # zero the parameter gradients
            optimizer.zero_grad()
    
            # forward + backward + optimize
            outputs = net(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
    
            # print statistics
            running_loss += loss.item()
            if i % 2000 == 1999:    # print every 2000 mini-batches
                print('[%d, %5d] loss: %.3f' %
                        (epoch + 1, i + 1, running_loss / 2000))
                running_loss = 0.0

print('Finished Training')
````

![Training your python model 2-2.png](<attachment:Training your python model 2-2.png>)
![Training your python model 3.png](<attachment:Training your python model 3.png>)

````python
correct = 0
total = 0
with torch.no_grad():
    for data in testloader:
        images, labels = data
        outputs = net(images)
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

print('Accuracy of the network on the 10000 test images: %d %%' % (
    100 * correct / total))
````
